```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(37) # 使得每次运行得到的随机数都一样
```

```python
import cv2
image=cv2.imread('E:\PyProjects\DataSet\FireAI/chair.jpg')
# Sobel 算子进行图像边缘检测
sobel_h=cv2.Sobel(image,cv2.CV_64F,1,0,ksize=3)
sobel_v=cv2.Sobel(image,cv2.CV_64F,0,1,ksize=3)
plt.figure(13,figsize=(15,30))

plt.subplot(131)
plt.imshow(image,cmap='gray') # 彩色图像显示异常，plt采用RGB模式，而cv2采用BGR模式
plt.title('raw_img')

plt.subplot(132)
plt.imshow(sobel_h,cmap='gray')
plt.title('sobel_h')

plt.subplot(133)
plt.imshow(sobel_v,cmap='gray')
plt.title('sobel_v')
```

```python
# Laplacian算子进行图像边缘检测
lap=cv2.Laplacian(image, cv2.CV_64F)
plt.figure(12,figsize=(10,30))

plt.subplot(121)
plt.imshow(image,cmap='gray') # 彩色图像显示异常，plt采用RGB模式，而cv2采用BGR模式
plt.title('raw_img')

plt.subplot(122)
plt.imshow(lap,cmap='gray')
plt.title('Laplacian')
```

```python
# Canny算子
canny = cv2.Canny(image, 50, 240)
plt.figure(12,figsize=(10,30))

plt.subplot(121)
plt.imshow(image,cmap='gray') # 彩色图像显示异常，plt采用RGB模式，而cv2采用BGR模式
plt.title('raw_img')

plt.subplot(122)
plt.imshow(canny,cmap='gray')
plt.title('Canny')
```
